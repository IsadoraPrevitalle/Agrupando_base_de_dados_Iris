```python
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
from sklearn import metrics
```

```python
colunas = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'variety']
df = pd.read_csv('iris.csv', names=colunas, header=None)

print(df.head())
```

```python
colunas = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'variety']
df_class = pd.read_csv('iris.csv', names=colunas, header=None)

print(df_class.head())
```

```python
df_class.drop(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], axis=1, inplace=True)
df_class
```

```python
df_class["variety"] = df_class["variety"].replace({
    'versicolor': 0,
    'setosa': 1,
    'virginica': 2
})
```

```python
df_class
```

```python
df_KM = df
```

```python
df_HI = df
```

```python
df_1 = np.array(df.drop('variety', axis=1))
df_1
```

```python
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(df_1)
print(labels)
```

```python
df_KM['k-classes'] = kmeans.labels_
df_KM
```

```python
sb.pairplot(df_KM, hue='variety')
```

```python
sb.pairplot(df_KM, hue='k-classes')
```

```python
kmeans.labels_
```

```python
print('Accuracy: ', metrics.accuracy_score(df_class, kmeans.labels_))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_true = df_class
y_pred = kmeans.labels_

cm = confusion_matrix(y_true, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='viridis', values_format='d')
plt.title('Matriz de Confusão - KMeans')
plt.show()
```

Agrupamento Hierarquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
dendograma = dendrogram(linkage(df_1, method='ward'))
```

```python
hierarquico = AgglomerativeClustering(n_clusters=3, linkage='ward', affinity='euclidean')
```

```python
hierarquico = hierarquico.fit_predict(df_1)
```

```python
hierarquico
```

```python
df_HI
```

```python
df_HI.drop(['k-classes'], axis=1, inplace=True)
```

```python
df_HI['HI-classes'] = hierarquico
```

```python
df_HI
```

```python
sb.pairplot(df_HI, hue='HI-classes')
```

```python
print('Accuracy: ', metrics.accuracy_score(df_class, hierarquico))
```

```python
y_true = df_class
y_pred = hierarquico

cm = confusion_matrix(y_true, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_true))
disp.plot(cmap='viridis', values_format='d')
plt.title('Matriz de Confusão - Clustering Hierárquico')
plt.show()
```

DBSCAN

```python
df_DBS = df
df_DBS
```

```python
df_DBS.drop(['HI-classes'], axis=1, inplace=True)
```

```python
df_DBS
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(df_1)
print(labels)
```

```python
df_DBS['DBS-classes'] = dbscan.labels_
df_DBS
```

```python
sb.pairplot(df_DBS, hue='DBS-classes')
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(df_1, labels)
print(f"Índice de Silhueta: {silhouette_avg}")
```

```python
y_true = df_class
y_pred = labels

mask = y_pred != -1
y_true_filtered = y_true[mask]
y_pred_filtered = y_pred[mask]

cm = confusion_matrix(y_true_filtered, y_pred_filtered, labels=np.unique(y_true))

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_true))
disp.plot(cmap='viridis', values_format='d')
plt.title('Matriz de Confusão - Clustering')
plt.show()
```
